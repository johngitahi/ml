```python
# HIV data exploration
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("/kaggle/input/kenya-hiv-data-tested-positive/2024-Kenya-HIV.csv")
```

```python
df.head()
```

```python
df = df[["County", "Total"]]
df["County"] = df["County"].str.replace(" County", "", regex=True)
```

```python
df.head()
```

```python
plt.figure(figsize=(20, 10))
plt.barh(df["County"], df["Total"], height=0.7)

plt.xlabel("No. of HIV+ tests")
plt.ylabel("County")
plt.title("2024 HIV incidence across the counties")
plt.gca().invert_yaxis()
plt.xticks(rotation=90, fontsize=8)  # Rotate 90Â° & reduce font size

plt.savefig("hiv_incidence_kenya_2024.png", dpi=300, bbox_inches="tight")
```
